```python
import pandas as pd

def cargar_dataset(ruta_dataset: str) -> pd.DataFrame:
    try: 
        df = pd.read_csv(ruta_dataset, encoding="latin-1", sep=",")
        print("Las primeras 5 filas del archivo son:")
        print(df.head())
        return df
    except FileNotFoundError: 
        print("Error: No se encuentra el archivo solicitado.")
        return pd.DataFrame()

ruta_dataset = r"C:\Users\rortega\Desktop\Python\Curso 3_Curso Aplicaciones en ciencia de datos e inteligencia artificial\Clase 1_pendiente_individual\Actividad 1\ejemplo_data.csv"
df = cargar_dataset(ruta_dataset)

# Transformar los atributos "ID" y "Activo".
if not df.empty:
    try:
        df["ID"] = df["ID"].astype(int)  # Utilizando la funcion astype transforme el atributo "ID" a entero. 
        df["Activo"] = df["Activo"].astype(bool)  #Utilizando la funcion astype transforme "Activo" a booleano.
       
        print("\nTransformación completada.")
        print("\nTipos de datos después de la transformación:")
        print(df.dtypes)

        print("\nInformación del DataFrame después de la transformación:")
        print(df.info())
    except KeyError as e:
        print(f"Error: No se encuentra la columna especificada: {e}")
    except ValueError as e:
        print(f"Error en la conversión de datos: {e}")
else:
    print("No se pudo cargar el DataFrame.")
```

```python
#Convierta el atributo "unidades" a entero y "2016" a flotante.

import pandas as pd

def cargar_dataset(ruta_dataset: str) -> pd.DataFrame:
    try: 
        df = pd.read_csv(ruta_dataset, encoding="latin-1", sep=",")
        print("Las primeras 5 filas del archivo son:")
        print(df.head())
        return df
    except FileNotFoundError: 
        print("Error: No se encuentra el archivo solicitado.")
        return pd.DataFrame()

ruta_dataset = r"C:\Users\rortega\Desktop\Python\Curso 3_Curso Aplicaciones en ciencia de datos e inteligencia artificial\Clase 1_pendiente_individual\Actividad 1\ejemplo_data.csv"

df = cargar_dataset(ruta_dataset)

df["Unidades"] = pd.to_numeric(df["Unidades"], errors='coerce').fillna(0).astype(int)
df["2016"] = pd.to_numeric(df["2016"], errors='coerce').fillna(0.0).astype(float)

# Transformar los atributos "unidades" y "2016"
if not df.empty:
    try:
        df["Unidades"] = df["Unidades"].astype(int)  # "unidades" a entero
        df["2016"] = df["2016"].astype(float)  #  "2016" a flotante
        
        print("\nTransformación completada.")
        print("\nTipos de datos después de la transformación:")
        print(df.dtypes)

        print("\nInformación del DataFrame después de la transformación:")
        print(df.info())
    except KeyError as e:
        print(f"Error: No se encuentra la columna especificada: {e}")
    except ValueError as e:
        print(f"Error en la conversión de datos: {e}")
else:
    print("\nNo se pudo cargar el DataFrame.")
```

```python
#Otra forma de resolver. 
import pandas as pd
import numpy as np

# Crear datos aleatorios
flores = ['Rosas', 'Gladiolo', 'Girasol', 'Bugambilia', 'Tulipan']
diccionario_flores = {
    'Flor': np.random.choice(flores, 50),
    "Color": np.random.randint(1, 20, 50),
    "Belleza": np.random.randint(1, 20, 50),
    "Duracion": np.random.randint(1, 20, 50),
    "Tallo": np.random.randint(1, 20, 50)
}

# Crear el DataFrame
df3 = pd.DataFrame(diccionario_flores)

# Estadísticas descriptivas de tendencia central y dispersión
estadisticas = pd.DataFrame({
    'Media': df3[["Color", "Belleza", "Duracion", "Tallo"]].mean(),
    'Mediana': df3[["Color", "Belleza", "Duracion", "Tallo"]].median(),
    'Varianza': df3[["Color", "Belleza", "Duracion", "Tallo"]].var(),
    'Desviación Estándar': df3[["Color", "Belleza", "Duracion", "Tallo"]].std(),
    'Rango': df3[["Color", "Belleza", "Duracion", "Tallo"]].max() - df3[["Color", "Belleza", "Duracion", "Tallo"]].min(),
    'Percentil 25': df3[["Color", "Belleza", "Duracion", "Tallo"]].quantile(0.25),
    'Percentil 75': df3[["Color", "Belleza", "Duracion", "Tallo"]].quantile(0.75)
})

# Mostrar estadísticas
print(estadisticas)
```
